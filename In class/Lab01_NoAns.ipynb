### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
n = int(input("Nhập n: "))
so_luong = (n - 1) - (n - 1) // 2
print(f"Số lượng số k thỏa mãn là: {so_luong}")
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
import math
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"{s:.1f}")
else:
    raise ValueError("Khong phai 3 canh cua tam giac")
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
x = input("Nhập một ký tự chữ cái: ")
n_str = input("Nhập số vị trí cần dịch chuyển (n): ")
n = int(n_str) 
a = x.lower()
print(f"Ký tự sau khi chuyển về chữ thường: {a}")
goc = ord('a')
ma_moi = (ord(a) - goc + n) % 26 + goc
ket_qua = chr(ma_moi)
print(f"Ký tự '{a}' sau khi dịch {n} vị trí là: {ket_qua}")
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
n = input("Nhập một chữ cái viết hoa: ")
if len(n) != 1:
    raise ValueError("Lỗi: Đầu vào phải là một ký tự duy nhất.")
if not n.isalpha():
    raise ValueError(f"Lỗi: '{n}' không phải là một chữ cái trong bảng chữ cái.")
ket_qua = n.lower()
print(f" Chữ thường: {ket_qua}")
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
n = int(input("Nhập chiều cao n: "))
for i in range(1, n + 1):
    print('*' * i)
```
