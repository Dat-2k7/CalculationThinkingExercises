### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 
### Ví dụ: Ta nhập h = 4

```python
h = int(input("Nhập h: "))
for i in range(1, h + 1):
    print(" " * (h - i) + "*" * (2 * i - 1))
```

### Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a

```python
a = int(input("Nhập vào số nguyên dương a: "))
if a <= 0:
    print("Vui lòng nhập một số nguyên dương!")
else:
    print(f"Các ước dương của {a} là:")
    for i in range(1, a + 1):
        if a % i == 0:
            print(i, end=" ")
```

### Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a

```python
a = int(input("Nhập vào số nguyên dương a: "))
count = 0
if a > 0:
    for i in range(1, a + 1):
        if a % i == 0:
            count += 1
    print(f"Số lượng các ước dương của {a} là: {count}")
else:
    print("Vui lòng nhập số nguyên dương!")
```

### Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
min = a if a < b else b
print(f"Ước chung của {a} và {b}:", end=" ")
for i in range(1, min + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

### Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…

```python
a = int(input("Nhập a: "))
so_cp = False
for i in range(1, a + 1):
    if i * i == a:
        so_cp = True
        break
    if i * i > a: break
print("Là số chính phương" if so_cp else "Không phải số chính phương")
```

### Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
a = int(input("Nhập a: "))
tong_uoc = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc += i
print("Là số hoàn hảo" if tong_uoc == a else "Không phải số hoàn hảo")
```

### Bài 11: Nhập vào số nguyên a, nếu nhập số âm thì yêu cầu nhập lại cho đến khi người dùng nhập đúng số dương. Nếu người dùng nhập đúng số dương thì in ra “Bạn nhập đúng quy tắc” và dừng chương trình.

### Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print(f"k cần tìm là: {k} (Tổng S = {s})")
```

### Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
### 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = float(input("Nhập A: "))
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print(f"n nhỏ nhất là: {n}")
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
max = None
min = None
while True:
    x = int(input("Nhập số (nhập -1 để dừng): "))
    if x == -1: break
    if max is None or x > max: max= x
    if min is None or x < min: min = x

if max is not None:
    print(f"Max: {max}, Min: {min}")
```

### Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập n: "))
so_chan = 0
so_le = 0
if n == 0: 
    so_chan = 1
else:
    while  n > 0:
        chu_so = n % 10 
        if chu_so % 2 == 0:
            so_chan += 1
        else:
            so_le += 1
        n //= 10 
print(f"Số chữ số chẵn: {so_chan}")
print(f"Số chữ số lẻ: {so_le}")
```

### Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập n: "))
tong = 0
while n > 0:
    chu_so = n % 10
    tong += chu_so 
    n //= 10
print(f"Tổng các chữ số của {n} là: {tong}")
```

### Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không

```python
n = int(input("Nhập n: "))
is_3k = True
if n <= 0: is_3k = False
else:
    while n > 1:
        if n % 3 != 0:
            is_3k = False
            break
        n //= 3
print("Là dạng 3^k" if is_3k else "Không phải dạng 3^k")
```

### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
n = int(input())
f1 = 1
f2 = 1
max = 0 
while max < n:
    f3 = f1 + f2 
    max = f3
    if max > n:
        max = f2
        break
    f1 = f2
    f2 = f3
print(max)
```

### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n = int(input("Nhập n: "))
count_nt = 0
while n > 0:
    i = n % 10
    if i in (2, 3, 5, 7):
        count_nt += 1
    n //= 10
print(f"Số chữ số là số nguyên tố: {count_nt}")
```
