## W4A1 
### [SumOfNumBers] 
### Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
n = int(input("Nhập n: "))
sum = 0
for i in range(1, n + 1):
    sum += i

print(f"Tổng các số từ 1 đến {n} là: {sum}")
```

## W4A6
### "[SumPrimesInRange] 
### Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."

```python
a, b = map(int, input().split())
tong = 0
for i in range(a, b + 1):
    if i < 2: continue
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0: break
    else:
        tong += i
print(tong)
```

## W4A7 
### "[LargestPrimeFactor] 
### Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

```python
n = int(input())
res = 0
i = 2
while i * i <= n:
    while n % i == 0:
        res = i
        n //= i
    i += 1
if n > 1: res = n
print(res)
```

### [PerfectSquaresDistinctDigits] 
### Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
n = int(input())
so = False
i = 1
while i * i <= n:
    so_cp = i * i
    s = str(so_cp) 
    is_so = True
    for j in range(len(s)):
        for k in range(j + 1, len(s)):
            if s[j] == s[k]: 
                is_so = False
                break
        if not is_so:
            break
    if is_so:
        print(so_cp, end=" ")
        so = True
    i += 1
if not so:
    print("(no number)")
```

## W4A13
### [ClosePairNumber] 
### Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. Ví dụ 220 và 284 là một cặp số thân thiết 220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 110. Tổng của 11 ước này cho chúng ta 284 284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng vừa đúng bằng 220 Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))
tong_uoc_a = 0
for i in range(1, a):
    if a % i == 0:
        tong_uoc_a += i
tong_uoc_b = 0
for i in range(1, b):
    if b % i == 0:
        tong_uoc_b += i
if tong_uoc_a == b and tong_uoc_b == a:
    print("true")
else:
    print("false")
```

## W4A14
### [GCD] 
### Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n.

```python
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))
ucln = 1
for i in range(1, min(m, n) + 1):
    if m % i == 0 and n % i == 0:
        ucln = i
print(ucln)
```

### [ChickenAndDog] 
### "Vừa gà vừa chó, 
### Bó lại cho tròn, 
### Ba mươi sáu con, 
### Một trăm chân chẵn" 
### Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid".

```python
tong_con = int(input("Nhập tổng số con: "))
tong_chan = int(input("Nhập tổng số chân: "))
tim_thay = False
for ga in range(tong_con + 1):
    cho = tong_con - ga
    if (ga * 2 + cho * 4) == tong_chan:
        print(f"{ga} {cho}")
        tim_thay = True
        break
if not tim_thay:
    print("invalid")
```
