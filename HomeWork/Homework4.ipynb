### W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input("Nhập n: "))
if n <= 0:
    print("Không phải")
else:
    while n > 1:
        if n % 2 != 0:
            break
        n //= 2
    if n == 1:
        print("Là dạng 2^k")
    else:
        print("Không phải")
```

### W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input("Nhập n: "))
tong = 0
while n > 0:
    tong += n % 10
    n //= 10
print("Tổng:", tong)
```

### W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập n: "))
chan = 0
le = 0
while n > 0:
    if (n % 10) % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Chẵn:", chan, "| Lẻ:", le)
```

### W4A23 Nhập n. Cho S(k) = 1 + 2 + 3 + … + k . Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print("k lớn nhất là:", k)
```

### W4A24 Nhập vào A, tìm n nhỏ nhất sao cho  1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = float(input("Nhập A: "))
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print("n nhỏ nhất là:", n)
```

### W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max = None
min = None
while True:
    x = int(input("Nhập số: "))
    if x == -1:
        break
    if max is None or x > max:
        max = x
    if min is None or x < min:
        min = x
print("Max:", max, "| Min:", min)
```

### W4A27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng). Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input("Nhập chuỗi: ")
count = 0
chuoi = False
for ky_tu in s:
    if ky_tu != " ":
        if not chuoi:
            count += 1
            chuoi = True
    else:
        chuoi = False
print("Số từ:", count)
```

### W4A28 Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi. Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input("Nhập chuỗi: ")
chuoi = False
for i in range(len(s)):
    if s[i] != " ":
        print(s[i], end="")
        chuoi = True
    elif chuoi:
        break
```

### W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa

```python
s = input("Nhập chuỗi: ")
for ky_tu in s:
    if 'a' <= ky_tu <= 'z':
        print(ky_tu.upper(), end="")
    elif 'A' <= ky_tu <= 'Z':
        print(ky_tu.lower(), end="")
    else:
        print(ky_tu, end="")
```

### W4A30 Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính tổng 3 số nguyên đó

```python
s = input("Nhập chuỗi số cách nhau dấu phẩy: ")
s += ","
tong = 0
so = ""
for chuoi in s:
    if chuoi.isdigit():
        so += chuoi
    elif chuoi == "," and so != "":
        tong += int(so)
        so = ""
print("Tổng:", tong)
```

### W4A31 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
s = input("Nhập chuỗi: ")
chu_hoa = 0
chu_thuong = 0
so = 0
for char in s:
    if char.isupper():
        chu_hoa += 1
    elif char.islower():
        chu_thuong += 1
    elif char.isdigit():
        so += 1
print("Chữ hoa:", chu_hoa, "| Chữ thường:", chu_thuong, "| Số:", so)
```
