### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.


```python
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        if math.isclose(a**2 + b**2, c**2) or math.isclose(a**2 + c**2, b**2) or math.isclose(b**2 + c**2, a**2):
            print("Tam giác vuông cân")
        else:
            print("Tam giác cân")
    elif math.isclose(a**2 + b**2, c**2) or math.isclose(a**2 + c**2, b**2) or math.isclose(b**2 + c**2, a**2):
        print("Tam giác vuông")
    else:
        print("Tam giác thường")
    chu_vi = a + b + c
    p = chu_vi / 2 # Nửa chu vi
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Chu vi: {chu_vi}")
    print(f"Diện tích: {dien_tich}")
else:
    print("Khong la ba canh cua mot tam giac")
```

### Bài 8: Giải hệ phương trình tuyến tính: ax + by = m và cx + dy = n




```python
a, b, m, c, d, n = map(float, input().split())
#Quy tắc nghiệm Cramer
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    print(f"Hệ có nghiệm duy nhất: x = {Dx / D}, y = {Dy / D}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x_c = float(input("Nhập hoành độ tâm đường tròn: "))
y_c = float(input("Nhập tung độ tâm đường tròn: "))
R = float(input("Nhập bán kính đường tròn R: "))
x_a = float(input("Nhập hoành độ điểm A: "))
y_a = float(input("Nhập tung độ điểm A: "))
# Bình phương khoảng cách từ tâm đến A
khoang_cach_binh_phuong = (x_a - x_c)**2 + (y_a - y_c)**2
# So sánh với bình phương bán kính
if khoang_cach_binh_phuong == R**2:
    print("Điểm A thuộc đường tròn (nằm trên đường tròn).")
elif khoang_cach_binh_phuong < R**2:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```
